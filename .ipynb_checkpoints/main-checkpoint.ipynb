loop version

```python
train_tensors = []
valid_tensors = []
for i in range(10):
    path_no = path/'training'/str(i)
    valid_path = path/'testing'/str(i)
    im = torch.stack([tensor(Image.open(o)) for o in path_no.ls().sorted()])
    im = im.float()/255
    train_tensors.append(im)
    img = torch.stack([tensor(Image.open(o)) for o in valid_path.ls().sorted()])
    img = img.float()/255
    valid_tensors.append(img)
```

here train_tensors is a list of 3D tensors.
